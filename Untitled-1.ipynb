Utilizei a API do Spotify com Python e o R para fazer uma requisição e obter os dados das minhas músicas salvas. Pude acessar as informações sobre cada música, como ID, nome da música, dançabilidade, acústica, energia, entre outros atributos.

```python
df = pd.read_csv('df_f.csv')
df_track = pd.read_csv('df.csv')
head_df = df.head()
```

```python
print(df)
```

```python
var_selecionadas= df[['id', 'danceability', 'acousticness', 'energy',
                       'loudness', 'speechiness', 'liveness', 'instrumentalness', 'valence']]

# Exibição das variáveis selecionadas
print(var_selecionadas)
```

```python
summary = var_selecionadas.describe()
print(summary)
```

```python
# Cálculo da matriz de correlação
cor_jose = var_selecionadas.iloc[:, 1:].corr()

# Plot do gráfico circular
plt.figure(figsize=(10, 10))
sns.heatmap(cor_jose, annot=True, cmap='coolwarm', cbar=False)
plt.title('Matriz de Correlação')
plt.show()
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='energy', y='loudness', color='#112446', edgecolor='none', s=30)
plt.title('Energy e loudness')
plt.xlabel('Energia')
plt.ylabel('Volume')
plt.annotate('Spotify', xy=(0.5, -0.1), xycoords='axes fraction', ha='center', fontsize=12, color='gray')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='valence', y='danceability', color='#112446', edgecolor='none', s=30)
plt.title('Valence e danceability')
plt.xlabel('Valência')
plt.ylabel('Dançabilidade')
plt.annotate('Spotify', xy=(0.5, -0.1), xycoords='axes fraction', ha='center', fontsize=12, color='gray')
plt.show()
```

```python
sns.set(style='whitegrid', font_scale=1.2)

plt.figure(figsize=(10, 8))
sns.scatterplot(data=df, x='energy' , y= 'valence', color='#112446', edgecolor='none', s=30)
plt.title('Energy e valence')
plt.xlabel('Energia')
plt.ylabel('Valência')
plt.annotate('Spotify', xy=(0.5, -0.1), xycoords='axes fraction', ha='center', fontsize=12, color='gray')
plt.show()
```

```python
# Aplicação do PCA
pca = PCA()
pca.fit(var_selecionadas.iloc[:, 1:])

# Componente Principal (PC1)
componente_principal = pca.transform(var_selecionadas.iloc[:, 1:])[:, 0]

# Variável de Saída (Energia)
variavel_saida = var_selecionadas['energy']

# Regressão Linear usando PC1 para prever a Energia
regressao = LinearRegression()
regressao.fit(componente_principal.reshape(-1, 1), variavel_saida)

# Coeficientes da Regressão
coeficiente_intercepto = regressao.intercept_
coeficiente_pc1 = regressao.coef_[0]

# Exibição dos coeficientes
print('Coeficiente de Intercepto:', coeficiente_intercepto)
print('Coeficiente de PC1:', coeficiente_pc1)
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=componente_principal, y=variavel_saida)
plt.plot(componente_principal, regressao.predict(componente_principal.reshape(-1, 1)), color='red', linewidth=2)
plt.title('Regressão Linear')
plt.xlabel('PC1')
plt.ylabel('Energia')
plt.show()
```
